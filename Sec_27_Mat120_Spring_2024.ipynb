```python
import numpy as np
import matplotlib.pyplot as plt

def lotka_volterra_system(t, state):
    R, W = state
    dRdt = R * (0.08 - 0.001 * W)
    dWdt = W * (-0.02 + 0.00002 * R)
    return np.array([dRdt, dWdt])

def euler_method(f, t0, tf, dt, state0):
    t = np.arange(t0, tf, dt)
    n = len(t)
    state = np.zeros((n, len(state0)))
    state[0] = state0
    for i in range(1, n):
        state[i] = state[i-1] + dt * f(t[i-1], state[i-1])
    return t, state

R0 = 1000
W0 = 40
state0 = [R0, W0]
t0, tf, dt = 0, 100, 0.1


t, states = euler_method(lotka_volterra_system, t0, tf, dt, state0)


R = states[:, 0]
W = states[:, 1]


plt.figure(figsize=(14, 4))

# Plot W vs t
plt.subplot(1, 3, 1)
plt.plot(t, W, label='Wolves')
plt.xlabel('Time')
plt.ylabel('Wolf Population')
plt.title('Wolf Population vs Time')
plt.legend()

# Plot R vs t
plt.subplot(1, 3, 2)
plt.plot(t, R, label='Rabbits')
plt.xlabel('Time')
plt.ylabel('Rabbit Population')
plt.title('Rabbit Population vs Time')
plt.legend()

# Plot W vs R
plt.subplot(1, 3, 3)
plt.plot(R, W, label='Phase Plot')
plt.xlabel('Rabbit Population')
plt.ylabel('Wolf Population')
plt.title('Wolf vs Rabbit Population')
plt.legend()

plt.tight_layout()
plt.show()
```
