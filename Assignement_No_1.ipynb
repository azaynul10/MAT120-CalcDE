```python
import numpy as np
import matplotlib.pyplot as plt

# Function 1
x = np.linspace(-10, 10, 100)
y1 = -2*x**3 + 16*x**2 - 15*x
plt.figure(figsize=(10, 6))
plt.subplot(3, 2, 1)
plt.plot(x, y1)
plt.title(r'$y=-2x^3+16x^2-15x$')
plt.xlim([-10, 10])

# Function 2
x = np.linspace(np.pi, 5*np.pi, 100)
y2 = np.log(x)
plt.subplot(3, 2, 2)
plt.plot(x, y2)
plt.title(r'$y=\ln(x)$')
plt.xlim([np.pi, 5*np.pi])

# Function 3
x = np.linspace(-np.pi, np.pi, 100)
y3 = np.sin(2*x)
plt.subplot(3, 2, 3)
plt.plot(x, y3)
plt.title(r'$y=\sin(2x)$')
plt.xlim([-np.pi, np.pi])

# Function 4
x = np.linspace(0, 15, 100)
y4 = 2*np.cos(x) + 3*np.sin(x)
plt.subplot(3, 2, 4)
plt.plot(x, y4)
plt.title(r'$y=2\cos(x)+3\sin(x)$')
plt.xlim([0, 15])

# Function 5
x = np.linspace(-3*np.pi, 3*np.pi, 100)
y5 = np.tan(x/4)
plt.subplot(3, 2, 5)
plt.plot(x, y5)
plt.title(r'$y=\tan(x/4)$')
plt.xlim([-3*np.pi, 3*np.pi])

# Function 6
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y6 = 10*x**3 + 9*x**2 - 7*x - 5
plt.subplot(3, 2, 6)
plt.plot(x, y6)
plt.title(r'$y=10x^3+9x^2-7x-5$')
plt.xlim([-2*np.pi, 2*np.pi])

plt.tight_layout()
plt.show()
```

```python
from sympy import *
import sympy as sp
init_printing(pretty_print=True)
# Function 1
x = sp.symbols('x')
y1 = -2*x**3 + 16*x**2 - 15*x
dy1 = sp.diff(y1, x)
critical_points_1 = sp.solveset(dy1, x)
display('Critical points for y1:', critical_points_1)
interval = sp.Interval(-10, 10)
critical_points_1 = sp.Intersection(critical_points_1, interval)

d2y1 = sp.diff(dy1, x)
for pt in critical_points_1:
    if d2y1.subs(x, pt) < 0:
        display(f'Maximum at x={pt}: y1({pt}) = {y1.subs(x, pt)}')
    else:
        display(f'Minimum at x={pt}: y1({pt}) = {y1.subs(x, pt)}')


# Function 2
x = sp.symbols('x')
y2 = sp.log(x)
dy2 = sp.diff(y2, x)
critical_points_2 = sp.solveset(dy2, x)
display('Critical points for y2:', critical_points_2)
interval = sp.Interval(sp.pi, 5*sp.pi)
critical_points_2 = sp.Intersection(critical_points_2, interval)

d2y2 = sp.diff(dy2, x)
for pt in critical_points_2:
    if d2y2.subs(x, pt) < 0:
        display(f'Maximum at x={pt}: y2({pt}) = {y2.subs(x, pt)}')
    else:
        display(f'Minimum at x={pt}: y2({pt}) = {y2.subs(x, pt)}')
# Function 3
x = sp.symbols('x')
y3 = sp.sin(2*x)
dy3 = sp.diff(y3, x)
critical_points_3 = sp.solveset(dy3, x)

interval = sp.Interval(-sp.pi, sp.pi)
critical_points_3 = sp.Intersection(critical_points_3, interval)

display('Critical points for y3 within [-pi, pi]:', critical_points_3)

d2y3 = sp.diff(dy3, x)
for pt in critical_points_3:
    if d2y3.subs(x, pt) < 0:
        display(f'Maximum at x={pt.evalf()}: y3({pt.evalf()}) = {y3.subs(x, pt).evalf()}')
    elif d2y3.subs(x, pt) > 0:
        display(f'Minimum at x={pt.evalf()}: y3({pt.evalf()}) = {y3.subs(x, pt).evalf()}')

# Function 4
x = sp.symbols('x')
y4 = 2*sp.cos(x) + 3*sp.sin(x)
dy4 = sp.diff(y4, x)
critical_points_4 = sp.solveset(dy4, x)


interval = sp.Interval(0, 15)
critical_points_4 = sp.Intersection(critical_points_4, interval)

display('Critical points for y4 within [0, 15]:', critical_points_4)

d2y4 = sp.diff(dy4, x)
for pt in critical_points_4:
    if d2y4.subs(x, pt) < 0:
        display(f'Maximum at x={pt.evalf()}: y4({pt.evalf()}) = {y4.subs(x, pt).evalf()}')
    elif d2y4.subs(x, pt) > 0:
        display(f'Minimum at x={pt.evalf()}: y4({pt.evalf()}) = {y4.subs(x, pt).evalf()}')

# Function 5
x = sp.symbols('x')
y5 = sp.tan(x/4)
dy5 = sp.diff(y5, x)
critical_points_5 = sp.solveset(dy5, x)


interval = sp.Interval(-3*sp.pi, 3*sp.pi)
critical_points_5 = sp.Intersection(critical_points_5, interval)

display('Critical points for y5 within [-3pi, 3pi]:', critical_points_5)

d2y5 = sp.diff(dy5, x)
for pt in critical_points_5:
    if d2y5.subs(x, pt) < 0:
        display(f'Maximum at x={pt.evalf()}: y5({pt.evalf()}) = {y5.subs(x, pt).evalf()}')
    elif d2y5.subs(x, pt) > 0:
        display(f'Minimum at x={pt.evalf()}: y5({pt.evalf()}) = {y5.subs(x, pt).evalf()}')
# Function 6
x = sp.symbols('x')
y6 = 10*x**3 + 9*x**2 - 7*x - 5
dy6 = sp.diff(y6, x)
critical_points_6 = sp.solveset(dy6, x)


interval = sp.Interval(-2*sp.pi, 2*sp.pi)
critical_points_6 = sp.Intersection(critical_points_6, interval)

display('Critical points for y5 within [-2pi, 2pi]:', critical_points_6)

d2y6 = sp.diff(dy6, x)
for pt in critical_points_6:
    if d2y6.subs(x, pt) < 0:
        display(f'Maximum at x={pt.evalf()}: y6({pt.evalf()}) = {y6.subs(x, pt).evalf()}')
    elif d2y6.subs(x, pt) > 0:
        display(f'Minimum at x={pt.evalf()}: y6({pt.evalf()}) = {y6.subs(x, pt).evalf()}')
```

```python
import sympy as sp
import numpy as np
from sympy import *
from scipy.integrate import quad

x, a, b = sp.symbols('x a b', positive=True)


f1 = sp.exp(-b*x) * sp.sin(a*x)
display('Indefinite integral of e^(-bx)*sin(ax):', sp.integrate(f1, x))


f2a = (a - sp.cos(x))**2
display('Integral of (a-cos(x))^2 from 0 to 2pi:', sp.integrate(f2a, (x, 0, 2*sp.pi)))


f2b = sp.exp(-sp.cos(x))
display('Integral of e^(-cos(x)) from -10 to 10:', sp.integrate(f2b, (x, -10, 10)))


f3 = x * sp.exp(6*x)
display('Integral of x*e^(6x) from -1 to 2:', sp.integrate(f3, (x, -1, 2)))

def f1_lambda(x, a, b):
    return np.exp(-b*x) * np.sin(a*x)

def f2_lambda(x, a):
    return (a - np.cos(x))**2

def f3_lambda(x):
    return np.exp(-np.cos(x))

def f4_lambda(x):
    return x * np.exp(6*x)

a, b = 1, 2
display('Numerical integral of e^(-bx)*sin(ax):', quad(f1_lambda, 0, np.inf, args=(a, b))[0])
display('Numerical integral of (a-cos(x))^2 from 0 to 2pi:', quad(f2_lambda, 0, 2*np.pi, args=(a))[0])
display('Numerical integral of e^(-cos(x)) from -10 to 10:', quad(f3_lambda, -10, 10)[0])
display('Numerical integral of x*e^(6x) from -1 to 2:', quad(f4_lambda, -1, 2)[0])
```
