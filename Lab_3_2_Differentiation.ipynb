## Differentiation

```python
#Make a symbol
x= symbols('x')

#Write symbolic expression
f = 2*x**3+ x**2 + 3*x+ 5

#taking the first derivative of the function
df = diff(f,x)

#taking the nth derivative of the function (passing the third argument)
df2 = diff(f,x,2)


# diff( "function" , "differentiate with rest to", "order of diff" )
# in case of first derivative, we do not need to specificy 1 as the order
# the derivative is stored in df, df2

# df3 = diff(f,x,3) means differentiating the function 3 times

#printing
print(df)

print(df2)

#Substituting values into expressions
print(f.subs(x,4))

print(df.subs(x,4))

#For evaluating a numerical expression into a floating point number use "evalf"
print(df.subs(x,4).evalf())
```

**The lambdify function (Coverting SymPy expression to a numerical expression)**

```python
x = symbols('x')
f = 2*x**3+ x**2 + 3*x+ 5
df = diff(f,x)

print(f)
print(df)

f = lambdify(x, f) #converting SymPy expression to a numerical expression

df = lambdify(x, df)


print(f(3))   #77

print(df(3))     #63

# expressions are always stored in variables, but using lambdify, allows us to turn them into numverical expressions
# thus we can print f(x) instead of f
```

# Maxima & Minima of a function

1. For a given function f(x), to find the critical/turning points, we solve f'(x)=0.

  The value of first derivative / f'(x) gives us the gradient at a certain point on the function, and it is always zero at critical points.

2. By equating f'(x), we obtain the x-coordinates of the critical points. These points are substituted in f(x) to find the y-coordinates. So now we have (x,y) which is a critical point. But is it a minima or a maxima?

3. We use the second derivative and check whether it is > or < or = 0.
  * x = c will be point of **local maxima** if f”(c) < 0
  * x = c will be point of **local minima** if f”(c) > 0
  * if f”(c) = 0, then the results will be declared **inconclusive**



In short:

Maxima : f'(x) = 0 and f”(c) < 0

Minima : f'(x) = 0 and f”(c) > 0

f''(x) = 0 means the test has failed, the results are inconclusive

Plotting $f(x)$ and $f'(x)$ in a given range

```python
from matplotlib.pyplot import *
import numpy as np
```

```python
# creating a function to take input the mathematical function and the range for plotting
def dplt(f, r):
  a = np.linspace(r[0], r[1], 1001)   #linspace( start, end, number of points )

  f_val = [f.subs(x,i) for i in a]    #creates a list of y values for each value of x in a
  g = diff(f,x)
  g_val = [g.subs(x,i) for i in a]    #creates a list of first derivative values for each value of x in a

  str1 = "function= "+ str(f)
  str2 = "derivative= "+ str(g)

  plot(a, f_val, label = str1 )
  plot(a, g_val, label = str2 )


  grid()
  legend()
  show()
```

```python
f = eval(input())
dplt(f, [-10, 10])
```

```python
# The code is not generalized

#creating a function which takes an equation as input, and prints out the maxima and minima
def extremum(f):
  df = diff(f, x)       #first derivative
  print(df)
  ddf = diff(df, x)     #second derivative       ddf = diff(f,x,2)
  print(ddf)
  crit_point = solve(df)      #finding turning points
  print(crit_point)

  crit_points = [i.evalf() for i in crit_point]      #creating floating point
  print(crit_points)
  f_val = [f.subs([(x,i)]).evalf() for i in crit_points]       #sub values of x in the eqn to find y values
  print(f_val)
  ddf_val = [ddf.subs([(x,i)]).evalf() for i in crit_points]       #sub values of x in the eqn to find whether its +ve or -ve
  print(ddf_val)
  for i in range(len(crit_points)):
    if ddf_val[i] < 0:
      print(f"at x={crit_points[i]}, we have local maxima, f({crit_points[i]})= {f_val[i]}")
    elif ddf_val[i] > 0:
      print(f"at x={crit_points[i]}, we have local minima, f({crit_points[i]})= {f_val[i]}")
    else:
      print("Pari Na ;( ")

# if you want to print only a certain number of digits, make sure to mention your number "n" as evalf(n)
```

# Extremum in an Interval

```python
def extremumr(f, r):
  df = diff(f, x)
  print(df)
  ddf = diff(df, x)
  print(ddf)
  crit_point = solveset(df, x,r)
  print(crit_point)
  crit_points = [i.evalf() for i in crit_point]
  print(crit_points)
  f_val = [f.subs([(x,i)]).evalf() for i in crit_points]
  print(f_val)
  ddf_val = [ddf.subs([(x,i)]).evalf() for i in crit_points]
  print(ddf_val)
  for i in range(len(crit_points)):
    if ddf_val[i] < 0:
      print(f"at x={crit_points[i]}, we have local maxima, f({crit_points[i]})= {f_val[i]}")
    elif ddf_val[i] > 0:
      print(f"at x={crit_points[i]}, we have local minima, f({crit_points[i]})= {f_val[i]}")
    else:
      print("Pari Na :( )")
```
