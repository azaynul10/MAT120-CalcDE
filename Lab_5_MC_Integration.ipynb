## An illustration of Monte Carlo integration:
* In this example, **the domain D is the inner circle** and **the domain E is the square**
* The square's area, $2^2=4$ & the area of the circle $π * 1^2$ can be easily calculated
* If we check the ratio:
$$ (π * 1 * 1) / (4) = 0.7854 $$
* Using the MC method, the area of the circle can be estimated using the ratio (0.8), given by ratio of points inside the circle (40) to the total number of points (50)

* This yields an approximation for the circle's area:
$$ \frac{\textit{Area of Circle}}{\textit{Area of Square}} \approx \frac{\textit{No. of dots in Circle}}{\textit{No. of dots in Square}} $$

$$ {\textit{Area of Circle}} \approx \frac{\textit{No. of dots in Circle} * {\textit{Area of Square}}}{\textit{No. of dots in Square}} $$

$$ {\textit{Area of Circle}} \approx 4*\frac{40}{50} = 3.2 ≈ π $$

## $\int_{0}^{1} e^x dx$

```python
x = smp.symbols('x')
smp.integrate(smp.exp(x), (x,0,1)).evalf()
```

```python
N = 9999          #large value will give more accurate result costing more computation time

x = np.random.rand(N)     # generating 9999 random x coordinates between 0 to 1

temp = np.random.rand(N)     # generating 9999 more points between 0 to 1
y = [np.e*i for i in temp]  # every point in temp, is used to evaluate e*temp, now we have a list of e*temp

points = [(x[i], y[i]) for i in range(N)]
# creating a list of pseudo-random-coordinates
# each element of x array and y array are paired up

#for i in range(10):
#  print(points[i])
# if you want to check the coordinates just print the above for loop

z = [np.e**i for i in x]

n = 0 #counts the number of points below the e**x line

for i in range(N):
  if y[i] <= z[i]:  # if the y coordinate is below the e**x, counting x-xoordinate isnt necessary since there is no whitespace on either sides
    n = n + 1

print(np.e*n/N)   #multiplying with exponential bc the area of the box = exp * 1
```

* Value will change every time you run this, since the coordinates are randomly generated in each run
* THe higher the number of points, the lower the error

## Evaluating $\pi$ using
$$ \pi = 4 \int_{0}^{1} \sqrt{1-x^2} dx $$

```python
N = 99999  #large value will give more accurate result costing more computation time
from numpy import sqrt

x = np.random.rand(N)
y = np.random.rand(N)


z = np.sqrt(1-x**2)

n = 0
for i in range(N):
  if y[i] <= z[i]:
    n = n + 1
print(4*n/N)   #multiplying with exponential bc the area of the box = 1 * 1, then the integral is multiplied by 4
```

```python
plt.scatter(x,y,color='red')
a = np.linspace(0,1,150)
plt.plot(a,np.sqrt(1-a**2))
```

## $\int_{0}^{1} x^x dx$

```python
j=np.linspace(0,1,150)
plt.plot(j,j**j)
```

```python
import numpy as np

f1 = lambda x: x**x
N = 9999
x = np.random.rand(N)
y = np.random.rand(N)

y1 = [f1(i) for i in x]
n=0

for i in range(N):
  if y[i] <= y1[i] :
    n += 1

integral_approximation = n / N     #multiplying with nothing, since the area of the box = 1 * 1

print("Approximation of the integral:", integral_approximation)
```

```python
1/math.e
```

```python
import math
import random

def function (x):
  return math.log(x) / x

x_points = []
y_points = []

count = 0          # count = total no. of points generated in the area
in_area = 0        # no. of points that will fall under the curve

while count < 1000:
  x_coord = random.uniform(1, 10)  # random.uniform(a, b) gives you a random floating point number in the range [a, b]
  y_coord = random.uniform(0, 1/math.e)   #highest point of the box we're making
  x_points.append([x_coord])
  y_points.append([y_coord])

  if y_coord < function(x_coord):
    in_area += 1                 # increasing in_area because the random point falls below the curve

  count += 1                     # increasing count because the total no. of random points in being counted & compared one by one, as this is a while loop

area_box = 9 / math.e

print((in_area / count)* area_box)
```

```python
x=np.linspace(1,10,2000)
plt.plot(x, np.log(x)/x,'o',color='r')
plt.plot([1.0,1.0],[0.0,1/math.e],color='g')
plt.plot([1.0,10.0],[1/math.e,1/math.e],color='g')
plt.plot([10.0,10.0],[0.0,1/math.e],color='g')
plt.plot([1.0,10.0],[0.0,0.0],color='g')
plt.scatter(x_points,y_points)
```
